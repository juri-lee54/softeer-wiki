```python
cur.execute("SELECT * FROM Customers LIMIT 3")

rows = cur.fetchall()
for row in rows:
    print(row)
```

- SQLite에서 지원 안 함

# SQL GROUP BY Statement


```python
cur.execute("""
SELECT Country, COUNT(CustomerID) AS [Number of Customers]
FROM Customers
GROUP BY Country;
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

```python
cur.execute("""
SELECT Country, COUNT(CustomerID) AS [Number of Customers]
FROM Customers
GROUP BY Country
ORDER BY COUNT(CustomerID) DESC;
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

## GROUP BY With JOIN Example


```python
cur.execute("""
SELECT Shippers.ShipperName, COUNT(Orders.OrderID) AS NumberOfOrders
FROM Orders
LEFT JOIN Shippers
ON Orders.ShipperID = Shippers.ShipperID
GROUP BY ShipperName;
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

# SQL HAVING Clause


```python
cur.execute("""
SELECT Country, COUNT(CustomerID) AS [Number of Customers]
FROM Customers
GROUP BY Country
HAVING COUNT(CustomerID) > 5;
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

```python
cur.execute("""
SELECT Country, COUNT(CustomerID) AS [Number of Customers]
FROM Customers
GROUP BY Country
HAVING COUNT(CustomerID) > 5
ORDER BY COUNT(CustomerID) DESC;
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

## More HAVING Examples


```python
cur.execute("""
SELECT Employees.LastName, COUNT(Orders.OrderID) AS NumberOfOrders
FROM Orders
INNER JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID
GROUP BY LastName
HAVING COUNT(Orders.OrderID) > 10;
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

```python
cur.execute("""
SELECT Employees.LastName, COUNT(Orders.OrderID) AS NumberOfOrders
FROM Orders
INNER JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID
WHERE LastName = 'Davolio' OR LastName = 'Fuller'
GROUP BY LastName
HAVING COUNT(Orders.OrderID) > 25;
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

# SQL Stored Procedures
- SQLite에서 지원 안 함

```python
cur.execute("""
CREATE PROCEDURE GetCustomersByCity
  @City nvarchar(50)
AS
BEGIN
  SELECT * FROM Customers
  WHERE City = @City;
END;
""")


cur.execute("""
EXEC GetCustomersByCity @City = 'London';
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```

## Stored Procedure With Multiple Parameters


```python
cur.execute("""
CREATE PROCEDURE GetCustomersByCity
  @City nvarchar(50),
  @PostalCode nvarchar(10)
AS
BEGIN
  SELECT * FROM Customers
  WHERE City = @City AND PostalCode = @PostalCode;
END;
""")


cur.execute("""
EXEC GetCustomersByCity @City = 'London', @PostalCode = 'WA1 1DP';
""")

rows = cur.fetchall()
for row in rows:
    print(row)
```
